```python
df_users["Location"] = df_users["Location"].replace('.*(?=\,)', '', regex=True)
df_users["Location"] = df_users["Location"].replace('\,\s', '', regex=True)
```

```python
df_users["Location"] = df_users["Location"].replace('\W' , '', regex=True)
df_users["Location"] = df_users["Location"].str.strip()
location_count = df_users['Location'].value_counts()
location_count_single = df_users['Location'].isin(location_count.index[location_count<=29])
df_users.loc[location_count_single, 'Location'] = "other"
df_users["Location"] = df_users["Location"].replace('' , 'other')
```

```python
df_books.loc[df_books["Book-Title"].str.contains("Harry Potter and the Chamber of Secrets")]
```

```python
for i in df_books.loc[df_books["Book-Title"].str.contains("Harry Potter and the Chamber of Secrets")]["ISBN"]:
    print(len(df_ratings.loc[df_ratings["ISBN"] == i ]))
```

```python
df_ratings_desc = pd.DataFrame(df_ratings["ISBN"].value_counts()).reset_index()
df_ratings_desc.columns = ["ISBN", "Count"]
df_ratings_desc = pd.merge(df_ratings_desc, df_books, on='ISBN')
```

```python
duplicated = df_ratings_desc.duplicated(subset=["Book-Title", "Book-Author"])
```

```python
len(df_ratings['ISBN'].unique())
```

```python
import time

ts = time.time()

for i in range(0, len(duplicated)):
    if duplicated[i]:
        curr_ISBN = df_ratings_desc.loc[(df_ratings_desc["Book-Title"] == df_ratings_desc["Book-Title"][i]) & 
                                        (df_ratings_desc["Book-Author"] == df_ratings_desc["Book-Author"][i])].reset_index()
        df_ratings["ISBN"] = df_ratings["ISBN"].replace(df_ratings_desc["ISBN"][i], curr_ISBN["ISBN"][0])

print ("Complete, time taken: " + str(time.time() - ts) + " seconds")
```

```python
len(df_ratings['ISBN'].unique())
```

```python
df_ratings = df_ratings.drop_duplicates(subset=["User-ID", "ISBN"])
```

```python
df_books = df_books[df_books["ISBN"].isin(df_ratings['ISBN'].unique())]
df_users = df_users[df_users["User-ID"].isin(df_ratings['User-ID'].unique())]

df_ratings = df_ratings[df_ratings["ISBN"].isin(df_books['ISBN'].unique())]
df_ratings = df_ratings[df_ratings["User-ID"].isin(df_users['User-ID'].unique())]
```

```python
rating_user_counts = df_ratings['User-ID'].value_counts()
ISBN_counts = df_ratings['ISBN'].value_counts()

df_ratings = df_ratings[df_ratings['User-ID'].isin(rating_user_counts[rating_user_counts >= 15].index)]
df_ratings = df_ratings[df_ratings['User-ID'].isin(rating_user_counts[rating_user_counts <= 500].index)]

df_ratings = df_ratings[df_ratings['ISBN'].isin(ISBN_counts[ISBN_counts >= 30].index)]
```

```python
df_books = df_books[df_books["ISBN"].isin(df_ratings['ISBN'].unique())]
df_users = df_users[df_users["User-ID"].isin(df_ratings['User-ID'].unique())]

df_ratings = df_ratings[df_ratings["ISBN"].isin(df_books['ISBN'].unique())]
df_ratings = df_ratings[df_ratings["User-ID"].isin(df_users['User-ID'].unique())]
```
